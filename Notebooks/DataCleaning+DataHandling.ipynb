```python
import pandas as pd
import numpy as np

# Create a working copy
churn_df = df.copy()

print("="*60)
print("TELCO CUSTOMER CHURN ANALYSIS - DATA CLEANING")
print("="*60)

# 1. Check churn column
print("\n1️ CHURN COLUMN ANALYSIS")
print("-"*40)
print(f"Churn column values: {churn_df['Churn'].unique()}")
print(f"Value counts:\n{churn_df['Churn'].value_counts()}")

# Convert Churn to binary (Yes=1, No=0)
churn_df['churn_binary'] = churn_df['Churn'].map({'Yes': 1, 'No': 0})
print(f"\n Converted 'Yes' → 1 (churned), 'No' → 0 (retained)")
print(f"Churn rate: {churn_df['churn_binary'].mean()*100:.1f}%")

# 2. Check customerID column
print("\n2️ CUSTOMER ID VERIFICATION")
print("-"*40)
print(f"Unique customers: {churn_df['customerID'].nunique()}")
print(f"Total rows: {len(churn_df)}")

if churn_df['customerID'].duplicated().sum() > 0:
    print(f" Warning: {churn_df['customerID'].duplicated().sum()} duplicate customer IDs found")
else:
    print(" All customer IDs are unique")
    
# 3. Check for missing values
print("\n MISSING VALUES CHECK")
print("-"*40)
missing = churn_df.isnull().sum()
missing_pct = (missing / len(churn_df)) * 100
missing_df = pd.DataFrame({'Missing': missing, 'Percentage': missing_pct})
missing_df = missing_df[missing_df['Missing'] > 0]

if len(missing_df) > 0:
    print(missing_df)
else:
    print(" No missing values found in any column")

# 4. Check TotalCharges column (often has issues)
print("\n4️ TOTALCHARGES COLUMN CHECK")
print("-"*40)
print(f"TotalCharges data type: {churn_df['TotalCharges'].dtype}")

# Convert TotalCharges to numeric (it might be object type with spaces)
churn_df['TotalCharges'] = pd.to_numeric(churn_df['TotalCharges'], errors='coerce')
print(f"After conversion: {churn_df['TotalCharges'].dtype}")

# Check for any conversion issues
missing_charges = churn_df['TotalCharges'].isnull().sum()
if missing_charges > 0:
    print(f" Found {missing_charges} rows with invalid TotalCharges")
    print(f"These will need to be filled or removed")

# 5. Handle duplicates
print("\n5️ DUPLICATE CHECK")
print("-"*40)
duplicates = churn_df.duplicated().sum()
if duplicates > 0:
    print(f"Found {duplicates} duplicate rows")
    churn_df = churn_df.drop_duplicates()
    print(f"Removed duplicates. New shape: {churn_df.shape}")
else:
    print(" No duplicate rows found")

# 6. Convert SeniorCitizen to categorical
print("\n6️ DATA TYPE OPTIMIZATION")
print("-"*40)
churn_df['SeniorCitizen'] = churn_df['SeniorCitizen'].astype('category')
print("Converted SeniorCitizen to category")

# Convert binary Yes/No columns to category
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for col in binary_cols:
    if col in churn_df.columns:
        churn_df[col] = churn_df[col].astype('category')
        print(f"Converted '{col}' to category")

# 7. Create tenure groups for better analysis
print("\n7️ CREATING TENURE GROUPS")
print("-"*40)
churn_df['tenure_group'] = pd.cut(churn_df['tenure'], 
                                   bins=[0, 12, 24, 48, 60, 72], 
                                   labels=['0-1 year', '1-2 years', '2-4 years', '4-5 years', '5+ years'])

print("Tenure groups created:")
print(churn_df['tenure_group'].value_counts())

# 8. Final data quality report
print("\n" + "="*60)
print("FINAL DATA QUALITY REPORT")
print("="*60)
print(f" Dataset shape: {churn_df.shape[0]} rows, {churn_df.shape[1]} columns")
print(f" Churn rate: {churn_df['churn_binary'].mean()*100:.1f}%")
print(f" Unique customers: {churn_df['customerID'].nunique()}")
print(f" Missing values: {churn_df.isnull().sum().sum()}")
print(f" Memory usage: {churn_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Display first few rows
print("\n CLEANED DATA PREVIEW:")
print(churn_df[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Churn', 'churn_binary']].head())

# Save cleaned data
churn_df.to_csv('telco_churn_cleaned.csv', index=False)
print("\n Cleaned data saved to: telco_churn_cleaned.csv")

print("="*60)
print("DATA CLEANING COMPLETE!")
print("="*60)
```

```python
print("="*60)
print("HANDLING MISSING TOTALCHARGES VALUES")
print("="*60)

# Check the rows with missing TotalCharges
missing_charges = churn_df[churn_df['TotalCharges'].isnull()]
print(f"\nRows with missing TotalCharges: {len(missing_charges)}")
print("\nThese customers:")
print(missing_charges[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']])

# Since TotalCharges = MonthlyCharges * tenure (approximately)
# Fill missing values with MonthlyCharges * tenure
churn_df['TotalCharges'] = churn_df['TotalCharges'].fillna(
    churn_df['MonthlyCharges'] * churn_df['tenure']
)

# Verify no more missing values
print("\n After filling:")
print(f"Missing values in TotalCharges: {churn_df['TotalCharges'].isnull().sum()}")

# Show the fixed rows
print("\nFixed rows:")
print(churn_df[churn_df['customerID'].isin(missing_charges['customerID'])][
    ['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']
])

print("\n" + "="*60)
print(" MISSING VALUES HANDLED!")
print("="*60)
```
